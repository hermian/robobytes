```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/XRPUSDT.csv')
df
```

```python
##### 단순 이동평균 계산

from ta.trend import SMAIndicator
df['sma7'] = SMAIndicator(df['c'], window=7).sma_indicator()
df['sma25'] = SMAIndicator(df['c'], window=25).sma_indicator()
df['sma99'] = SMAIndicator(df['c'], window=99).sma_indicator()
df.head(10)
```

```python
##### 가중 이동평균 계산

from ta.trend import WMAIndicator
df['wma7'] = WMAIndicator(df['c'], window=7).wma()
df['wma25'] = WMAIndicator(df['c'], window=25).wma()
df['wma99'] = WMAIndicator(df['c'], window=99).wma()
df.head(10)
```

```python
##### 지수 이동평균 계산

from ta.trend import EMAIndicator
df['ema7'] = EMAIndicator(df['c'], window=7).ema_indicator()
df['ema25'] = EMAIndicator(df['c'], window=25).ema_indicator()
df['ema99'] = EMAIndicator(df['c'], window=99).ema_indicator()
df.head(10)
```

```python
##### MACD 계산

from ta.trend import MACD
macd = MACD(df['c'], window_slow=26, window_fast=12, window_sign=9)
df['macd'] = macd.macd()
df['macd_s'] = macd.macd_signal()
df['macd_d'] = macd.macd_diff()
df.tail(10)
```

```python
##### RSI 계산

from ta.momentum import RSIIndicator
df['rsi6'] = RSIIndicator(df['c'], window=6).rsi()
df['rsi12'] = RSIIndicator(df['c'], window=12).rsi()
df['rsi24'] = RSIIndicator(df['c'], window=24).rsi()
df.head(10)
```

```python
##### StochRSI 계산

from ta.momentum import StochRSIIndicator
stochRSI = StochRSIIndicator(df['c'], window=14, smooth1=3, smooth2=3)
df['srsi'] = stochRSI.stochrsi()
df['srsik'] = stochRSI.stochrsi_k()
df['srsid'] = stochRSI.stochrsi_d()
df.tail(10)
```

```python
##### Bollinger Bands 계산

from ta.volatility import BollingerBands
bb = BollingerBands(df['c'], window=20, window_dev=2)
df['bh'] = bb.bollinger_hband() #high band
df['bhi'] = bb.bollinger_hband_indicator() #high band 보다 가격이 높으면 1, 아니면 0
df['bl'] = bb.bollinger_lband() #low band
df['bli'] = bb.bollinger_lband_indicator() #low band 보다 가격이 낮으면 1, 아니면 0
df['bm'] = bb.bollinger_mavg() #middle band
df['bw'] = bb.bollinger_wband() #band width
df.tail(10)
```

```python
##### VWAP 계산

from ta.volume import VolumeWeightedAveragePrice
vwap = VolumeWeightedAveragePrice(high=df['h'], low=df['l'], close=df['c'], volume=df['v'], window=14)
df['vwap7'] = vwap7.volume_weighted_average_price()
df.tail(10)
```

```python
df = df.dropna()
df.columns
```

```python
df.to_csv('./data/XRPUSDT_index.csv', index=False)
```
